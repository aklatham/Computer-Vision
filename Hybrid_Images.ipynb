Cross Correlation

```
def cross_correlation_2d(filt, image):

  fh = np.shape(filt)[0]
  fw = np.shape(filt)[1]
  ih = np.shape(image)[0]
  iw = np.shape(image)[1]
  ic = np.shape(image)[2]
  fb1 = fh//2 + 1
  fb2 = fw//2 + 1

  ret = np.zeros(np.shape(image))

  work = np.zeros((fb1 + ih, fb2 + iw, ic))
  work[fb1:fb1+ih, fb2:fb2+iw] = image
  filt = np.reshape(filt, -1)

  for i in range(iw):
    for j in range(ih):
      cross = np.reshape(work[j:j+fh, i:i+fw], ((fh * fw), ic))
      ret[j, i] = np.dot(filt, cross)

  return ret.astype(np.uint8)
```

Convolve

```
def convolve_2d(filt, image):
  filt = np.flip(filt, 0)
  filt = np.flip(filt, 1)

  ret = cross_correlation_2d(filt, image)
  return ret
```

Gaussian Blur

```
def gaussian_blur_kernel_2d(sigma, size):

  ret = np.zeros((size, size))

  def gaussian(x, y):
      return (1 / (2 * np.pi * sigma) * np.exp(-1 * ((np.power(x, 2) + np.power(y, 2)) / (2 * sigma))))

  for x in range(-int(size/2), size - int(size/2)):
        for y in range(-int(size/2), size -int(size/2)):
            ret[x + int(size/2), y + int(size/2)] = gaussian(x, y)

  return ret
```

Low Pass

```
def low_pass(image, sigma):
  filt = gaussian_blur_kernel_2d(sigma, 3)
  ret = convolve_2d(filt, image)
  return ret
```

High Pass

```
def high_pass(image, sigma):
  filt = gaussian_blur_kernel_2d(sigma, 3)
  low = convolve_2d(filt, image)
  ret = image - low
  return ret
```

Hybrid Constructor

```
def hybrid_constructor(imageh, imagel, sigma):
  low = low_pass(imagel, sigma)
  high = high_pass(imageh, sigma)
  ret = low + high
  return ret
```
